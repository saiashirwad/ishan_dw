## Suburb
In this section, the coordinates of all the properties are used to locate the suburbs they are a part of. A polygon is to be found for each property such that the coordinates lie within its perimeter.

```python
# Suburb
shp = shapefile.Reader('vic_suburb_boundary/VIC_LOCALITY_POLYGON_shp.dbf')
all_shapes = shp.shapes() 
all_records = shp.records()     

def get_suburb(inp):
    try:
        property_id, point = inp
        for i in range(len(all_shapes)):
            boundary = all_shapes[i] 
            if Point(point).within(shape(boundary)):
                return (property_id, all_records[i].as_dict()['VIC_LOCA_2'])
        return ("", "")
    except:
        pass
```

```python
# Locating the suburb a property is situated in 
if 'prop2sub.json' not in os.listdir("."):
    lat = list(real_state.lat)
    lng = list(real_state.lng)
    coords = [(i, j) for i, j in zip(lng, lat)]
    property_ids = list(real_state.property_id)
    inp = [(pid, coord) for pid, coord in zip(property_ids, coords)]
    
    with Pool(4) as p:
        property2suburb = p.map(get_suburb, inp)
else:
    with open('prop2sub.json', 'r') as f:
        property2suburb = json.load(f)
    prop2sub = {}
    for i in property2suburb:
        try:
            prop2sub[i[0]] = i[1]
        except:
            pass
```

```python
def in_weekdays(service_id):
    temp = 0
    for i in ["monday", "tuesday", "wednesday", "thursday", "friday"]: 
        temp += calendar[calendar.service_id==service_id][i].values[0]
    if temp > 0:
        return True
    return False
```

```python
def travel_min_to_CBD(stop_id):
    times = []
    ts = list(stop_times_7_9[stop_times_7_9.stop_id == stop_id].trip_id.unique())
    # Ignoring all trips that don't occur on weekdays
    for t in ts:
        sid = trips[trips.trip_id == t].service_id.values[0]
        if not in_weekdays(sid):
            ts.remove(t)
            
    for t in ts:
        try:
            st_time = stop_times_7_9[(stop_times_7_9.trip_id == t) & (stop_times_7_9.stop_id == stop_id)].departure_time
            re_time = stop_times_7_9[(stop_times_7_9.trip_id == t) & (stop_times_7_9.stop_id == 19854)].arrival_time
            re_time = pd.to_datetime(re_time, format='%H:%M:%S').values[0]
            st_time = pd.to_datetime(st_time, format='%H:%M:%S').values[0]
            times.append(pd.Timedelta(re_time - st_time).seconds / 60)
        except Exception as e:
            pass
    try:
        return (stop_id, round(sum(times) / len(times), 2))
    except:
        return (stop_id, None)
```

```python
%%time
stop_ids = list(stops.stop_id.unique())
with Pool(4) as p:
    stopid2cbd = p.map(travel_min_to_CBD, stop_ids)
```

This function has been written to get a quick overview of the dataset before and after applying several transformations. It computes the mean and standard deviation, plots chosen columns next to each other, and trains a linear regression on the model with only those columns considered and outputs its score. The goal is not to train a good model with it but to see if the data has the potential to be fit well by a linear model.

```python
def describe(df, columns=columns): 
    for column in columns:
        mean = np.mean(list(df[column]))
        std = np.std(list(df[column]))
        print(column)
        print("Mean: " , mean, " Standard Deviation: ", std)
    
    sns.pairplot(df[columns])
    
    X = df[X_cols]
    y = df['price']
    
    model = linear_model.LinearRegression()
    model.fit(X, y)
    print("Score: ", model.score(X, y))
    
```

## Standardization

Just like it was with the unmodified data, standardization of all the columns does not yield any real benefits when it comes to performing well with a linear model.The same is observed when a MinMax transformation was applied for both. 

```python
standard_scaler = StandardScaler()
real_state_standard = real_state.copy()
real_state_standard[columns] = standard_scaler.fit_transform(real_state_standard[columns])
```

```python
describe(real_state_standard, columns)
```

```python
minmax_scaler = MinMaxScaler(feature_range=(0, 1))
real_state_minmax = real_state.copy()
real_state_minmax[X_cols] = standard_scaler.fit_transform(real_state_minmax[X_cols])
describe(real_state_minmax, columns)
```

It is observed that a log transformation allows the data to fit a linear regression model much better. 

```python
log_normalized = real_state.copy()
for index in log_normalized.index:
    for column in columns:
        val = log_normalized.loc[index, column]
        try:
            log_normalized.loc[index, column] = np.log(val + 1)
        except Exception as e:
            pass
```

```python
describe(log_normalized, columns)
```

From the following figures, it is visible that a box-cox transformation results in the data in the columns assuming a bell curve more than any other transformation or normalization method so far. A clear inverse relationship is observed between travel_min_to_CBD and price. This transformation does not yield a score as good as the log transformation, but the figures are a lot more compelling, and this could potentially lead to a model with less overfitting.

```python
real_state_boxcox = real_state.copy()
pt = PowerTransformer(method='box-cox')
normalizer = Normalizer()

real_state_boxcox[columns] = normalizer.fit_transform(real_state_boxcox[columns])
describe(real_state_boxcox, columns)
```
